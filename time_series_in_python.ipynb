```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
adf = adfuller(x=dataset['Adj Close'])
```

```python
def test(value):
    if value > 0.05:
        print(f'we fail to reject the null hypothesis. the dataset is not stationary with p-value: {value} ')
    else:
        print(f'we reject null hypothesis. the dataset is stationary with p-value: {value}')
```

```python
test(adf[1])
```

```python
dataset_diff = np.diff(a=dataset['Adj Close'], n=1)
```

```python
adf_diff = adfuller(x=dataset_diff)
test(adf_diff[1])
```

```python
plot_acf(dataset_diff, lags=20)
plt.show()
```

# moving average process

```python
sales = pd.read_csv('data/widget_sales.csv')
sales.head()
```

```python
sales.info()
```

```python
date = pd.date_range(start='01-01-2019', periods=500)
```

```python
sales.index = date
sales.head()
```

```python
fig, ax = plt.subplots(figsize=(16, 5),
                       dpi=300)
sales.plot(ax=ax)
plt.title('widget sales')
plt.show()
```

```python
sales_test = RandomWalk(sales.widget_sales)
```

```python
sales_diff = pd.DataFrame({'sales': sales_test.himpunan})
sales_diff.head()
```

```python
split = int(len(sales_diff)*0.9)
train = sales_diff[:split]
test = sales_diff[split:]
```

```python
def RollingForecast(df: pd.DataFrame,
                    train_len: int,
                    horizon: int,
                    window: int,
                    method: str):
    total_len =  train_len + horizon
    if method == 'mean':
        pred_mean = []
        for i in range(train_len,
                       total_len, 
                       window):
            mean = np.mean(df[:i].values)
            pred_mean.extend(mean for _ in range(window))
        return pred_mean

    elif method == 'last':
        pred_last_value = []
        for i in range(train_len,
                       total_len,
                       window):
            last_value = df[:i].iloc[-1].values[0]
            pred_last_value.extend(last_value for _ in range(window))
        return pred_last_value

    elif method == 'MA':
        pred_MA = []
        for i in range(train_len,
                       total_len,
                       window):
            model = SARIMAX(df[:i],
                            order=(0,0,2))
            res = model.fit(disp=False)
            predictions = res.get_prediction(0, i + window - 1)
            oos_pred = predictions.predicted_mean.iloc[-window:]
            pred_MA.extend(oos_pred)
        return pred_MA
```

```python
pred_df = test.copy()

TRAIN_LEN = len(train)
HORIZON = len(test)
WINDOW = 2

pred_mean = RollingForecast(sales_diff, TRAIN_LEN, HORIZON, WINDOW, 'mean')
pred_last = RollingForecast(sales_diff, TRAIN_LEN, HORIZON, WINDOW, 'last')
pred_MA = RollingForecast(sales_diff, TRAIN_LEN, HORIZON, WINDOW, 'MA')

pred_df['pred_mean'] = pred_mean
pred_df['pred_last'] = pred_last
pred_df['pred_MA'] = pred_MA

pred_df.head()
```

```python
pred_df.plot(figsize=(16,5))
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
for i in pred_df:
    print(i)
    print(f"r2_score: {round(r2_score(pred_df['sales'], pred_df[i]),3)}")
    print(f"mean_squared_error: {round(mean_squared_error(pred_df['sales'], pred_df[i]), 3)}")
    print("")
```
