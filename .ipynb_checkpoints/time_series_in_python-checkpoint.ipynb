```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf
```

```python
dataset = pd.read_csv('data/GOOGL.csv')
dataset.head()
```

```python
dataset['Date'] = pd.to_datetime(dataset['Date'])
dataset.info()
```

```python
dataset = dataset[['Date', 'Adj Close']]
dataset.head()
```

```python
dataset.index = dataset['Date']
dataset = dataset.drop('Date', axis=1)
dataset.head()
```

```python
fig, ax = plt.subplots(figsize=(16,5))
dataset.plot(ax=ax)
plt.title('GOOGL stock')
plt.ylabel('close stock in usd')
plt.show()
```

```python
adf = adfuller(x=dataset['Adj Close'])
```

```python
def test(value):
    if value > 0.05:
        print(f'we fail to reject the null hypothesis. the dataset is not stationary with p-value: {value} ')
    else:
        print(f'we reject null hypothesis. the dataset is stationary with p-value: {value}')
```

```python
test(adf[1])
```

```python
dataset_diff = np.diff(a=dataset['Adj Close'], n=1)
```

```python
adf_diff = adfuller(x=dataset_diff)
test(adf_diff[1])
```

```python
plot_acf(dataset_diff, lags=20)
plt.show()
```
